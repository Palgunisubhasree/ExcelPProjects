```python
import os
os.listdir('/content/Dataset_IRIS')
```

```python
import pandas as pd

df = pd.read_csv('/content/Dataset_IRIS/iris.data',
                 header=None,
                 names=['sepal_length','sepal_width','petal_length','petal_width','species'])

df.head()
```

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

df['cluster'] = kmeans.labels_
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['petal_length'], df['petal_width'], c=df['cluster'])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("K-Means Clustering on Iris Dataset")
plt.show()
```

3. Description of the Dataset

The Iris dataset contains 150 samples of iris flowers with 4 numerical features:

Sepal Length

Sepal Width

Petal Length

Petal Width

It includes three species: Iris-setosa, Iris-versicolor, and Iris-virginica.
This dataset is commonly used for clustering and classification.

4. Python Code to Load Dataset and Apply K-Means
import pandas as pd

# Load dataset (.data file without headers)
df = pd.read_csv('/content/Dataset_IRIS/iris.data',
                 header=None,
                 names=['sepal_length','sepal_width','petal_length','petal_width','species'])

df.head()

Apply K-Means Clustering
from sklearn.cluster import KMeans

# Select only numeric features
X = df[['sepal_length','sepal_width','petal_length','petal_width']]

# Apply K-Means with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Add cluster labels to the dataset
df['cluster'] = kmeans.labels_
df.head()

5. Visualization (Scatter Plot)
import matplotlib.pyplot as plt

plt.scatter(df['petal_length'], df['petal_width'], c=df['cluster'])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("K-Means Clustering on Iris Dataset")
plt.show()

6. Interpretation

K-Means with K=3 successfully grouped the dataset into three clusters.

One cluster corresponds almost entirely to Iris-setosa, which has very distinct petal measurements.

The remaining two clusters represent Iris-versicolor and Iris-virginica, which have overlapping features but still show visible separation.

The scatter plot shows three clear groups based on petal length and width.
